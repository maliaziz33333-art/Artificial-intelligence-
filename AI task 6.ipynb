```python
1. BFS without Queue & without Node
```

```python
def bfs_simple(graph, start_node):
    visited = []
    queue = [start_node]

    while queue:
       
        current = queue.pop(0)
        
        if current not in visited:
            visited.append(current)
            # Add neighbors to the end of the list
            neighbors = graph[current]
            for n in neighbors:
                queue.append(n)
                
    return visited

my_graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print(bfs_simple(my_graph, 'A'))
```

```python
2. BFS with Queue & Node
```

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def bfs_with_queue(root_node):
    if not root_node:
        return []

    results = []
    queue = deque([root_node])

    while queue:
    
        current = queue.popleft()
        results.append(current.value)

        # Check children
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
            
    return results

root = Node("Home")
root.left = Node("Kitchen")
root.right = Node("Bedroom")
root.left.left = Node("Pantry")

print(bfs_with_queue(root))
```
