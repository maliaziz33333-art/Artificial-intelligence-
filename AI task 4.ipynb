```python
1. Code for LUHN Algorithm
```

```python
def check_luhn(card_num):
    digits = [int(d) for d in str(card_num)]
    
    for i in range(len(digits) - 2, -1, -2):
        doubled_val = digits[i] * 2
        if doubled_val > 9:
            digits[i] = doubled_val - 9
        else:
            digits[i] = doubled_val
            
    total = sum(digits)
    return total % 10 == 0

test_card = "79927398713"
print(check_luhn(test_card))
```

```python
2. Write a python program to remove punctuations from the given string?
```

```python
def remove_punc(text_input):
    symbols = "!()-[]{};:'\"\,<>./?@#$%^&*_~"
    clean_text = ""
    
    for char in text_input:
        if char not in symbols:
            clean_text += char
            
    return clean_text

example = "Hello, world! How's it going?"
print(remove_punc(example))
```

```python
3. Write a python program to sort the sentence in alphabetical order?
```

```python
def sort_sentence(sentence):
    words = sentence.split()
    words.sort(key=str.lower)
    
    result = " ".join(words)
    return result

my_string = "Python is great for artificial intelligence"
print(sort_sentence(my_string))
```
