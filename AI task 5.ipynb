```python
1. DFS with Stack & Node
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def dfs_with_stack(root):
    if root is None:
        return []

    stack = [root]
    visited = []

    while stack:
        current = stack.pop()
        visited.append(current.value)

        if current.right:
            stack.append(current.right)
        if current.left:
            stack.append(current.left)
            
    return visited


tree_root = Node(1)
tree_root.left = Node(2)
tree_root.right = Node(3)
tree_root.left.left = Node(4)
tree_root.left.right = Node(5)

print(dfs_with_stack(tree_root))
```

```python
2. Research about "Inorder, Preorder, Postorder" and implement in DFS
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def preorder(node):
    if node:
        print(node.val, end=" ")
        preorder(node.left)
        preorder(node.right)

def inorder(node):
    if node:
        inorder(node.left)
        print(node.val, end=" ")
        inorder(node.right)

def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.val, end=" ")


root = Node("A")
root.left = Node("B")
root.right = Node("C")

print("Preorder:")
preorder(root)
print("\nInorder:")
inorder(root)
print("\nPostorder:")
postorder(root)
```
